```python
df2 = df2.astype(str)

df2["phrase_najdi"] = df2["phrase_najdi"].str.strip()
df2["msa_meaning"] = df2["msa_meaning"].str.strip()
df2["English"] = df2["English"].str.strip()
```

```python
# إنشاء نسخة للتدريب
train_df = df2.copy()

# صياغة الإدخال (Prompt)
train_df["input"] = (
    "Translate Najdi Arabic to MSA and English.\n"
    "Najdi: " + train_df["phrase_najdi"]
)

# صياغة المخرجات (Target)
train_df["output"] = (
    "MSA: " + train_df["msa_meaning"] + "\n"
    "English: " + train_df["English"]
)

# نحتفظ فقط بأعمدة التدريب
train_df = train_df[["input", "output"]]

train_df.head()
```

```python
from transformers import EarlyStoppingCallback

collator = DataCollatorForSeq2Seq(tokenizer, model)

args = TrainingArguments(
    output_dir="./najdi_en_model",
    eval_strategy="epoch",
    save_strategy="epoch",

    # أهم إضافات:
    load_best_model_at_end=True,
    metric_for_best_model="eval_loss",
    greater_is_better=False,
    save_total_limit=2,          # مهم عشان لا يمتلئ الدسك

    learning_rate=2e-5,          # أقل من قبل
    num_train_epochs=15,         # بدل 40
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    weight_decay=0.01,
    warmup_ratio=0.1,

    fp16=torch.cuda.is_available(),
    report_to=[]
)

trainer = Trainer(
    model=model,
    args=args,
    train_dataset=train_tok,
    eval_dataset=val_tok,
    data_collator=collator,
    callbacks=[EarlyStoppingCallback(early_stopping_patience=5)]
)

trainer.train()
```

```python
import torch

def predict_najdi_to_en(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True).to(model.device)

    with torch.no_grad():
        out = model.generate(
            **inputs,
            num_beams=5,
            max_new_tokens=40,
            repetition_penalty=1.2,
            no_repeat_ngram_size=3,
            early_stopping=True
        )

    return tokenizer.decode(out[0], skip_special_tokens=True).strip()

print(predict_najdi_to_en("يعطيك العافية"))
print(predict_najdi_to_en("وش علومك"))
print(predict_najdi_to_en("عساك بخير"))
```

```python
import torch

device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

def translate_najdi_to_en(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=64).to(device)
    with torch.no_grad():
        out = model.generate(
            **inputs,
            max_new_tokens=60,
            num_beams=5,
            no_repeat_ngram_size=3,
            early_stopping=True
        )
    return tokenizer.decode(out[0], skip_special_tokens=True).strip()

tests = ["خل عنك", "ربي يتمم", "ما علينا "]
for t in tests:
    print("Najdi:", t)
    print("EN   :", translate_najdi_to_en(t))
    print("-"*30)
```
